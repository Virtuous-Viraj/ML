```python
# string based data handling like male and female
df["Gender"] = "Female"
df.iloc[0:10, 6] = "Male"
df.head()
```

```python
def f(s):
    if s == "Male":
        return 0
    return 1
df["sex"] = df.Gender.apply(f)
df.head()
```
