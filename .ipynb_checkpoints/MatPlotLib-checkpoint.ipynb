```python
# plot function x^3
import numpy as np
# x = np.array([1,2,3,4])
x = np.arange(0,5,0.1)
y = x**3
y2 = x**2

plt.plot(x,y, color="orange", marker=">")
plt.show()
```

```python
#color, marker, linewidth, labels for axis , datasets and the title
plt.plot(x,y, color="orange", linewidth="5", label="X^3")
plt.plot(x,y2, color="red", linewidth="5", label = "X^2")

plt.axis([0,5,0,100])
plt.grid()

plt.text(2,80, 'Text', fontsize=12)
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Matplotlib Demo")
plt.legend()
plt.show()
```

```python
## Bubble Chart
```

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = [100, 120, 180, 250, 300, 370]
```

```python
c = np.arange(len(year))
```

```python
plt.scatter(year, salary, s=population, c = c, edgecolor="red")
plt.text(year[0]+0.1,salary[0]+0.1, population[0])
plt.text(year[4]+0.1,salary[4]+0.1, population[4])
plt.show()
```

```python
## Pie Graphs
```

```python
sizes = [3,4,6,2]
labels = ["A", "B", "C", "D"]
colors = ["blue", "purple", "red", "pink"]
plt.title("")
plt.pie(sizes, colors = colors, labels = labels, autopct="%.2f%%")
plt.axis("equal")
plt.show()
```

```python
explode = [0.1, 0 , 0, 0]
plt.pie(sizes, colors = colors, labels = labels, autopct="%.2f%%", explode = explode, startangle = 100)
plt.axis("equal")
plt.show()
```
